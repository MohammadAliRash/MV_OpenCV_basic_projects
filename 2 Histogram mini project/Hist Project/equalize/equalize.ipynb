```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#####################################creating RGB image

#reading img
original = cv2.imread('2.jpg')

#chechikng
print(original.shape)

#showing
plt.imshow(original)
plt.title('original')
```

```python
#fixing
original_fixed = cv2.cvtColor(original,cv2.COLOR_BGR2RGB)

#cheching
print(original_fixed.shape)

#showing
plt.imshow(original_fixed)
plt.title('original_fixed')
plt.imsave('original_fixed.jpg', original_fixed)
```

```python
#if img is gray_scale
original_fixed_gray= cv2.cvtColor(original_fixed, cv2.COLOR_RGB2GRAY)

original_fixed_gray_equalized = cv2.equalizeHist(original_fixed_gray)

plt.imshow(original_fixed_gray_equalized , cmap='gray')
plt.title('original_fixed_gray_equalized')
plt.imsave('original_fixed_gray_equalized.jpg', original_fixed_gray_equalized)
```

```python
plt.imshow(original_fixed_gray , cmap='gray')
plt.title('original_fixed_gray')
plt.imsave('original_fixed_gray.jpg', original_fixed_gray)
```

```python
#if img is colored

# Convert to HSV colorspace
hsv = cv2.cvtColor(original_fixed, cv2.COLOR_BGR2HSV)
```

```python
# Grab V channel
hsv[:,:,2]
```

```python
hsv[:,:,2] = cv2.equalizeHist(hsv[:,:,2])
```

```python
# Convert back to RGB to visualize
original_fixed_equalized = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
plt.imshow(original_fixed_equalized)
```
